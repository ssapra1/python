```python
# prompt: pd.DataFrame(geo_encoder,one_hot_encoder.get_feature_names_out

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder
# pip install tensorflow #already installed in the previous code execution
# pip install ipykernel #already installed in the previous code execution

data=pd.read_csv('Churn_Modelling.csv')
#print(data.head()) # commented out as not required in the task

# pre process the data
#
data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
# encode one categorial variable
data1=data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
label_gender_encode=LabelEncoder()
data1['Gender']=label_gender_encode.fit_transform(data1['Gender'])
#print(data1) # commented out as not required in the task

# We have encoded the gender
# now Geography has categorial features
#
# we can encode Geography with Label encoder as it is not ggood but we will use one hot encoder

one_hot_encoder=OneHotEncoder(sparse_output=False)

geo_encoder=one_hot_encoder.fit_transform(data1[['Geography']])

geo_encoder_df=pd.DataFrame(geo_encoder,columns=one_hot_encoder.get_feature_names_out(['Geography']))

geo_encoder_df
```
